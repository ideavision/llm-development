## Setting up Hugging Face

### Create a Hugging Face account

1. Go to the [signup](https://huggingface.co/join) page.
2. Enter your email and password, click next and fill in profile details to create an account.

<img src="images/huggingface-join.png" width=200 />

4. Go to the [Access Tokens](https://huggingface.co/settings/tokens) page.
5. Create a new token to programmatically authenticate your identity to the Hugging Face Hub. Doing this allows you to read, create and update your public or private repositories on the Hub.

<img src="images/huggingface-new-token.png" width=200 />

6. Give a meaningful name to the token, and change the Role to "write", then click "Generate a token".

<img src="images/huggingface-new-token-generate.png" width=200 />

## Setting up Workspace Integrations

### Adding Environment variables to Workspace

1. In the Workspace left-hnd navigation bar, click on Environment.

<img src="images/workspace-environment.png" alt="Workspace left-hand navigation bar with Environment button highlighted." width="50">

2. Click on the "Add environment variables" plus button.

<img src="images/workspace-add-environment-variables.png" alt="Pkus button to add new environment variables" width="200">

3. Fill out the name and value fields as follows.

| Environment Variables Set Name     | Name              | Value             |
|------------------------------------|-------------------|------------------ |
|             HuggingFace            | HUGGINGFACE_TOKEN | _your secret key_ |

<img src="images/workspace-env-var-details.png" width="200">

5. Click the "Create" button.
6. In the Environment pane, in the list of environment variables, click "..." to open the context menu, then "Connect" to connect the new integration.

<img src="images/workspace-env-var-connect.png" alt="Connect environment variables context menu" width=200 />
